# Loading necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

# Loading train and test data 

```python
df_train_80 = pd.read_csv("bank_loan_train_80.csv")
df_train_90 = pd.read_csv("bank_loan_train_90.csv")
df_train_100 = pd.read_csv("bank_loan_train_100.csv")
df_test = pd.read_csv("bank_loan_test_data.csv")
```

# Logistic Regression

#### for upsampling ratio of 0.8

```python
lr_80 = LogisticRegression()
lr_80.fit(df_train_80.drop('Loan Status',axis=1),df_train_80['Loan Status'])
```

```python
print("The accuracy of logistic regression model with upsampling ratio 0.8 is {}".format(lr_80.score(df_test.drop('Loan Status',axis=1),df_test['Loan Status'])))
```

```python
y_pred = lr_80.predict(df_test.drop('Loan Status',axis=1))
cn_matrix = confusion_matrix(df_test['Loan Status'],y_pred)
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cn_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cn_matrix.flatten()/np.sum(cn_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2,v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cn_matrix, annot=labels, fmt="", cmap='Blues')
```

## Looking at value of TP,TN,FP,FN for different value of n_neighbors in KNN algorithm

```python
n_neig = [1,3,5,7,9,11,15,25,45]
knn_score = []
confusion_knn = []
confusion_percentage_knn = []
for i in n_neig:
    knn = KNeighborsClassifier(n_neighbors=i,algorithm='kd_tree')
    knn.fit(df_train_100.drop('Loan Status',axis=1),df_train_100['Loan Status'])
    accuracy = knn.score(df_test.drop('Loan Status',axis=1),df_test['Loan Status'])
    knn_score.append(accuracy)
    y_pred = knn.predict(df_test.drop('Loan Status',axis=1))
    matrix = sklearn.metrics.confusion_matrix(df_test['Loan Status'],y_pred).ravel()
    confusion_knn.append(matrix.ravel())
    per = (matrix.ravel()/np.sum(matrix))*100
    confusion_percentage_knn.append(per)
```

```python
confusion_knn = np.array(confusion_knn)
confusion_percentage_knn = np.array(confusion_percentage_knn)
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
for i in range(0,4):
    plt.plot(n_neig,confusion_knn[:,i],label=labels[i])
    plt.xlabel('Number of Neighbors')
    plt.ylabel('Number of sample')
    plt.legend(loc=1)
    plt.axis()
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
for i in range(0,4):
    plt.plot(n_neig,confusion_percentage_knn[:,i],label=labels[i])
    plt.xlabel('Number of Neighbors')
    plt.ylabel('percentage of sample')
    plt.legend(loc=1)
    plt.axis()
```

## Looking at value of TP,TN,FP,FN for different value of n_estimators in Random Forest algorithm

```python
n_tree = [10,20,30,50,70,90,150,200,250]
rf_score = []
confusion = []
confusion_percentage = []
for i in n_tree:
    rf = RandomForestClassifier(n_estimators=i)
    rf.fit(df_train_100.drop('Loan Status',axis=1),df_train_100['Loan Status'])
    accuracy = rf.score(df_test.drop('Loan Status',axis=1),df_test['Loan Status'])
    rf_score.append(accuracy)
    y_pred = rf.predict(df_test.drop('Loan Status',axis=1))
    matrix = sklearn.metrics.confusion_matrix(df_test['Loan Status'],y_pred).ravel()
    confusion.append(matrix.ravel())
    per = (matrix.ravel()/np.sum(matrix))*100
    confusion_percentage.append(per)
```

```python
confusion = np.array(confusion)
confusion_percentage = np.array(confusion_percentage)
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
for i in range(0,4):
    plt.plot(n_tree,confusion[:,i],label=labels[i])
    plt.xlabel('Number of Tree')
    plt.ylabel('Number of sample')
    plt.legend(loc=1)
    plt.axis()
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
for i in range(0,4):
    plt.plot(n_tree,confusion_percentage[:,i],label=labels[i])
    plt.xlabel('Number of Tree')
    plt.ylabel('percentage of sample')
    plt.legend(loc=1)
    plt.axis()
```
